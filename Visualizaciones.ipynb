```python
pictures = data['Nombre'].unique()
ns = data['N'].unique()

best_n = {}
for p in pictures:
    rows = data.loc[data['Nombre'] == p]
    r = rows['Tasa de compresión (porcentual)'].idxmax(axis=1)
    best_n[p]=rows.loc[r]['N']
    
n_counts={}
for n in ns:
    n_counts[n] = sum(value == n for value in best_n.values())
    
ns = pd.DataFrame.from_dict(n_counts, orient='index')
sns.barplot(x=ns.index, y=ns.loc[:,0],data=ns)
```

```python
# N boxplot
sns.catplot(x="N", y="Tasa de compresión (porcentual)",
                height=3.5, aspect=1.5,
                kind="box", legend=True, data=data);
```
